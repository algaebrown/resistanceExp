```python
# identify clusters
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

path = '/home/hermuba/data0118/network1122/'
def edge_list(fname):
    
    file = path + fname
    edge = pd.read_csv(file, header = 0)
    edge[['gene_one', 'gene_two']].to_csv(path + fname + '_edgelist', index = False, header = False, na_rep = 'nan')
def to_net(fname):
    # generate edge_list
    edge_list(fname)
    
    net = nx.read_edgelist(path + fname + '_edgelist', delimiter = ',', nodetype = str,)
    return(net)
file_e = 'refseq_ordinary_40_2'
net = to_net(file_e)
```

```python
gold_anno = pd.read_pickle('/home/hermuba/data0118/goldstandard/node_anno_df')
```

```python
gold_anno.head()
```

```python
pan = pd.read_csv('/home/hermuba/data0118/cdhit/clstr/pangenome_df/Escherichia0.70.clstr.csv', index_col = 0, header = 0)
```

```python
count = pan.merge(gold_anno[['cluster', 'GO']], left_index = True, right_on = 'cluster').drop(columns = ['GO', 'cluster'], axis = 1).sum(axis = 1)
```

```python
gold_anno['count'] = count
```

```python
gold_anno['count'].plot(kind = 'hist', bins = 100)
```

```python
attr = ['cluster', 'card', 'hypo_nr']
for a in attr:
    nx.classes.function.set_node_attributes(net,gold_anno[a].to_dict(), a)
nx.classes.function.set_node_attributes(net,count.to_dict(), 'count')
```

```python
net.nodes['562.22834.con.0008_97|562.22834']
```

```python
#degree centrality
dc = nx.algorithms.centrality.degree_centrality(net)
```

```python
for k in dc.keys():
    gold_anno.loc[k ,'degree_centrality']  = dc[k]
```

```python
gold_anno['degree_centrality'] = gold_anno['degree_centrality'].astype(float)
```

```python
gold_anno.plot(kind='scatter', x = 'count', y='degree_centrality', logy=True)
```

```python
gold_anno['is_card']=gold_anno['card'].notnull()
gold_anno.loc[gold_anno.loc[gold_anno['count'] >1500].index, 'core'] = True
gold_anno['core'] = gold_anno['core'].fillna(False)
```

```python
gold_anno.boxplot(by = 'is_card', column = 'degree_centrality')
```

```python
# betweeness centrality: slow
bc = nx.algorithms.centrality.betweenness_centrality(net)
for k in bc.keys():
    gold_anno.loc[k ,'between_centrality']  = bc[k]
gold_anno['between_centrality'] = gold_anno['between_centrality'].astype(float)
gold_anno.plot(kind='scatter', x = 'count', y='between_centrality', logy=True)
```
