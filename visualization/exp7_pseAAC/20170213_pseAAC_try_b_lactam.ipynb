```python
# 
def match_strain_pse(cluster_no):
    pseAAC = pd.read_pickle("/home/hermuba/data/genePredicted/cdhit/cluster_pseAAC/" + cluster_no)
    member = cluster_detail.loc['Cluster '+cluster_no, 'member']
    member = member.split(',')[1:]
    for gene in member:
        strain = gene.split('|')[1]
        gene_id = gene.split('|')[0]
        pseAAC.loc[gene_id, 'strain'] = strain
        pseAAC.to_pickle("/home/hermuba/data/genePredicted/cdhit/cluster_pseAAC/" + cluster_no)
```

```python
for i in lactamase:
    match_strain_pse(i.split(' ')[1])
```

```python
ris = pd.read_pickle("/home/hermuba/data/annotated_RIS/anno_sps_df")
ris = ris.loc[ris['Species'] == 'Escherichia']
```

```python
ris.groupby('Antibiotic').count()['index'].sort_values()
```

```python
from sklearn import linear_model
from sklearn.model_selection import LeaveOneOut
import numpy as np
to_binary={'Resistant': 1,'Susceptible': 0, 'Not defined': 0, 'Non-susceptible': 0,'Intermediate':0 }
def pre_cluster(abx):
    needed = ris.loc[ris['Antibiotic']==abx][['Genome ID', 'Resistant Phenotype']]
    df = abs_pattern.merge(needed, left_index = True, right_on = 'Genome ID')
    df.reset_index(inplace = True, drop = True)
    X = df[list(lactamase)+list(efflux_pump)+list(pbp)]
    y = df['Resistant Phenotype'].map(to_binary)
    return(X, y)
def pre_lact_cluster(abx):
    needed = ris.loc[ris['Antibiotic']==abx][['Genome ID', 'Resistant Phenotype']]
    df = abs_pattern.merge(needed, left_index = True, right_on = 'Genome ID')
    df.reset_index(inplace = True, drop = True)
    X = df[list(lactamase)]
    y = df['Resistant Phenotype'].map(to_binary)
    return(X, y)
def train(X,y):
    loo = LeaveOneOut()
    score_ma = 0
    for train_index, test_index in loo.split(X):

        X_train, X_test = X.loc[train_index], X.loc[test_index]
        y_train, y_test = y[train_index], y[test_index]
        reg = linear_model.LogisticRegression()
        reg.fit(X_train, y_train)
        if reg.predict(X_test)[0] == int(y_test):
            score_ma += 1
    
    return(score_ma/len(y))
def run(abx):
    X, y = pre_cluster(abx)
    s = train(X, y)
    return(s)
def run_lact(abx):
    X, y = pre_lact_cluster(abx)
    s = train(X, y)
    return(s)
```

```python
def add_pseAAC(feature_set):
    added = pd.DataFrame()
    for i in feature_set:
        pseAAC = pd.read_pickle("/home/hermuba/data/genePredicted/cdhit/cluster_pseAAC/" + i.split(' ')[1])
        #print(pseAAC)
        added = added.append(pseAAC)
    return(added)
def sum_pse(added):
    return(added.groupby('strain').sum())
def mean_pse(added):
    return(added.groupby('strain').mean())
sum_lact = sum_pse(add_pseAAC(lactamase))
mean_lact = mean_pse(add_pseAAC(lactamase))
def pre_sum(abx):
    needed = ris.loc[ris['Antibiotic']==abx][['Genome ID', 'Resistant Phenotype']]
    df = sum_lact.merge(needed, left_index = True, right_on = 'Genome ID')
    df.reset_index(inplace = True, drop = True)
    
    X = df.drop(columns = ['Resistant Phenotype', 'Genome ID'])
    y = df['Resistant Phenotype'].map(to_binary)
    return(X, y)
def pre_mean(abx):
    needed = ris.loc[ris['Antibiotic']==abx][['Genome ID', 'Resistant Phenotype']]
    df = mean_lact.merge(needed, left_index = True, right_on = 'Genome ID')
    df.reset_index(inplace = True, drop = True)
    X = df.drop(columns = ['Resistant Phenotype', 'Genome ID'])
    y = df['Resistant Phenotype'].map(to_binary)
    return(X, y)
```

```python
def run_sum(abx):
    X, y = pre_sum(abx)
    s = train(X, y)
    return(s)
```

```python
def run_mean(abx):
    X, y = pre_mean(abx)
    s = train(X, y)
    return(s)
```

```python
drug_of_interest = ['ampicillin', 'ampicillin/sulbactam', 'cefazolin', 'cefepime', 'ceftazidime','ceftriaxone', 'piperacillin/tazobactam', 'meropenem','imipenem', 'ertapenem', 'gentamicin', 'amikacin', 'ciprofloxacin', 'levofloxacin']

result = pd.DataFrame(columns = ['cluster', 'lactamase_cluster_only','sum_of_pseAAC', 'mean_of_pseAAC'])
for d in drug_of_interest:
    result.loc[d] = [run(d),run_lact(d), run_sum(d), run_mean(d)]
```

```python
result.plot(kind = 'bar', figsize=(10,5), legend = 'reverse')
```

```python
from sklearn import svm
def train_svm(X,y, k):
    loo = LeaveOneOut()
    score_ma = 0
    for train_index, test_index in loo.split(X):

        X_train, X_test = X.loc[train_index], X.loc[test_index]
        y_train, y_test = y[train_index], y[test_index]
        reg = svm.SVC(kernel=k, C=1)
        reg.fit(X_train, y_train)
        if reg.predict(X_test)[0] == int(y_test):
            score_ma += 1
    
    return(score_ma/len(y))
```

```python

def run_svm(abx, kernel, feature):
    X, y = feature(abx)
    s = train_svm(X, y, kernel)
    return(s)

result_svm_linear = pd.DataFrame(columns = ['cluster', 'sum_of_pseAAC', 'mean_of_pseAAC'])
for d in drug_of_interest:
    result_svm_linear.loc[d] = [run_svm(d, 'linear', pre_cluster), run_svm(d, 'linear', pre_sum), run_svm(d, 'linear', pre_mean)]
```

```python
result_svm_linear.plot(kind = 'bar', figsize=(10,5), legend = 'reverse')
```

```python
result_svm_rbf = pd.DataFrame(columns = ['cluster', 'sum_of_pseAAC', 'mean_of_pseAAC'])
for d in drug_of_interest:
    result_svm_rbf.loc[d] = [run_svm(d, 'rbf', pre_cluster), run_svm(d, 'rbf', pre_sum), run_svm(d, 'rbf', pre_mean)]
```

```python
result_svm_rbf.plot(kind = 'bar', figsize=(10,5), legend = 'reverse')
```

```python
r = result_svm_rbf.merge(result_svm_linear, left_index = True, right_index = True, suffixes= ['_rbf', '_linear'])
r = r.merge(result, left_index = True, right_index = True)
```
