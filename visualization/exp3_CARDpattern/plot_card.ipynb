```r
#load data
library(readxl)
df <- read_excel("../../ec_card.xlsx")

# sort by SIR
sorted <- df[with(df, order(cefazolin_ris)), ]
```

# cefazolin

```r
#aro from 3000010 to 3004045
start_aro = grep("3000010", colnames(df))
end_aro = grep("3004045", colnames(df))

# turn to matrix
ma <- data.matrix(sorted[,start_aro:end_aro])

# draw
library(gplots)
library(RColorBrewer)

hmcol = colorRampPalette(brewer.pal(9, "RdBu"))(100)
hmap <- heatmap.2(ma, labRow =sorted[,c('cefazolin_ris')], dendrogram = "row", Rowv = TRUE, Colv = TRUE, scale="none", trace = "none", col = hmcol, breaks = seq(-3, 3, length.out = 101))
```

```r
PCA <- prcomp(ma)
plot(PCA$x[,1:2])
text(PCA$x[,1], PCA$x[,2], sorted[,c('cefazolin_ris')], cex=0.6, pos=4, col="red") 
```
