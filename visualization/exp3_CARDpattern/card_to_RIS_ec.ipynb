```python
import pandas as pd
combined = pd.read_pickle('../../../combined')
combined.head()

# integrate to new RIS anootated
anno_RIS = pd.read_excel('../../EC_amr_anotateRIS20170710.xlsx')
anno_RIS.head()

# remove space in categoricals
for i in list(combined.columns.values):
    if '_ris' in str(i):
        combined[i] = combined[i].cat.rename_categories(['R','I','S'])
        combined[i] = combined[i].cat.add_categories(['cannot determine'])
```

```python
# mapping
for index, row in anno_RIS.iterrows():
    antibiotic = row['ANTIBIOTIC'] + '_ris'
    #print(combined.ix[row['GENOME_ID'],antibiotic], row['SIR'])
    combined.ix[row['GENOME_ID'],antibiotic] = row['SIR']
```

```python
# all E.coli
ecoli_card = combined.loc[list(set(list(anno_RIS['GENOME_ID']))),:]
```
