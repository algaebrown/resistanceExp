```python
# sort by classes of drug
drug_df = pd.read_pickle('/home/hermuba/data0118/drug_category')
```

```python
drug_df.head()
```

```python
drug_df = drug_df.loc[drug_df['drug'].isin(count_ris_df.index)]
```

```python
drug_order = drug_df.sort_values(by = 'order')['drug']
```

```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('seaborn-white')

def abx_abundance_plot(ris_distribution):
    # sort by resistance
    ris_distribution = ris_distribution[['Susceptible', 'Intermediate', 'Resistant']]
    
    # transpose it
    ris_distribution = ris_distribution.transpose()
    
    # sort by drug class
    ris_distribution = ris_distribution[drug_order]
    
    # create subplot
    f, axarr = plt.subplots(3, sharex=True, figsize = (10,7))
    
    
    
    # plot heatmap
    axarr[0].pcolor(ris_distribution, cmap = 'Blues')
    axarr[0].set_title('Phenotype Data Distribution: Count')
    #f.colorbar(orientation='vertical', ax = axarr[0])
    
    # set y tick for first plot
    plt.sca(axarr[0])
    plt.yticks(np.arange(0.5, ris_distribution.shape[0], 1), ris_distribution.index)

   
    # plot heatmap: frequency
    freq = ris_distribution/ris_distribution.sum(axis = 'index')
    axarr[1].pcolor(freq, cmap = 'Blues')
    #f.colorbar(orientation='vertical', ax = axarr[1])
    axarr[1].set_title('Phenotype Data Distribution: Frequency (%)')
    
    # set y tick for second plot
    plt.sca(axarr[1])
    plt.yticks(np.arange(0.5, ris_distribution.shape[0], 1), ris_distribution.index)


    x = np.arange(0.5, ris_distribution.shape[1], 1)
    axarr[2].bar(x, ris_distribution.sum(axis = 'index'))
    axarr[2].set_title('Number of data points')
    
    
    # set 
    plt.tight_layout()
    plt.sca(axarr[2])
    #plt.ylabel('No. Data', rotation=0)
    plt.xlabel('Drugs')
    plt.xticks(np.arange(0.5, ris_distribution.shape[1], 1), ris_distribution.columns)
    plt.setp(plt.xticks()[1], rotation=90)
    plt.grid(b=None)
    
    plt.savefig('/home/hermuba/data0118/fig/abx_data_abundance_heatmap.pdf', format = 'pdf', dpi = 300, pad_inches = 1, bbox_inches='tight')
    #
    return(plt)
```

```python
p = abx_abundance_plot(count_ris_df)
```
