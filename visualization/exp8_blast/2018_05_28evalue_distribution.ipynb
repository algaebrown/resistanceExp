```python
import pandas as pd
df = pd.read_pickle("/home/hermuba/data0118/blast_test") #blast E.coli0.70 representing genes against local nr, no evalue limit
df.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt
log_evalue = df['evalue'].apply(np.log).replace([np.inf, -np.inf], np.nan).dropna()
plt.hist(log_evalue, bins = 1000)
```

```python
hypothetical = df.loc[df['stitle'].str.contains('hypothetical protein')]
log_e_hypo = hypothetical['evalue'].apply(np.log).replace([np.inf, -np.inf], np.nan).dropna()
plt.hist(log_e_hypo, bins = 1000)
```
