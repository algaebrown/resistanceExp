```python
# extract only one
first_go = df.groupby(by = 'qseqid')['goterm'].first()
last_go = df.groupby(by = 'qseqid')['goterm'].last()
```

```python
# drop rows without goterms
no_na_df = df.dropna(axis = 'index', subset = ['goterm'])
# join the go terms together
go_term_sum = no_na_df.groupby(by = 'qseqid')['goterm'].apply(lambda x: "%s" % '|'.join(x))
# to remove redundant go terms from concatenation
no_redun_go = go_term_sum.apply(lambda x: set(x.split('|')))
```

```python
no_redun_go
```

```python
total_list = []
for i in no_redun_go:
    total_list += list(i)
counting = pd.Series(total_list)
```

```python
%matplotlib inline
counting.value_counts()[:100].plot(kind = 'bar', figsize = (20,2))
```
