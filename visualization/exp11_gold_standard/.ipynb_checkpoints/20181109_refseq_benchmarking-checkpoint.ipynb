```python
%matplotlib inline
test_data = '/home/hermuba/data0118/refseq_mu_test.csv'
import pandas as pd
d = pd.read_csv(test_data, header = 0)
d['goldstandard'] = d['goldstandard'].map({'f':False, 't':True})
d.groupby(by = 'goldstandard')[['mutual_info', 'nrm_mutual']].describe()
```

```python
d.boxplot(column = ['mutual_info', 'nrm_mutual'], by = 'goldstandard')
```

```python
d.groupby(by = 'goldstandard')['mutual_info'].plot(kind = 'kde', legend = True)
```

```python
d.groupby(by = 'goldstandard')['nrm_mutual'].plot(kind = 'kde', legend = True)
```

```python
%matplotlib inline
test_data = '/home/hermuba/data0118/refseq_mu_test_not_null.csv'
import pandas as pd
d = pd.read_csv(test_data, header = 0)
d['goldstandard'] = d['goldstandard'].map({'f':False, 't':True})
d.groupby(by = 'goldstandard')[['mutual_info', 'nrm_mutual']].describe()
```

```python
d.groupby(by = 'goldstandard')['mutual_info'].plot(kind = 'kde', legend = True)
```

```python
d.groupby(by = 'goldstandard')['nrm_mutual'].plot(kind = 'kde', legend = True)
```

```python
d['cut_mutual'] = pd.cut(d['nrm_mutual'], 100)
grouped = d.groupby(by = ['goldstandard', 'cut_mutual']).count()['gene_one']
lls = grouped[True]/grouped[False]/(grouped[True].sum()/grouped[False].sum())
lls.plot()
```
