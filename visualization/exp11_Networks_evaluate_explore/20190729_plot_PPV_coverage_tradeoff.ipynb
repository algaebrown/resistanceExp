```python
# plot together!!
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
plt.style.use('seaborn-white')
# import all PPV-coverage tradeoff
domain_tradeoff = pd.read_csv('~/data0118/PPV_coverage/domain_rm_plasmid_250', header = 0)
eskape_tradeoff = pd.read_csv('~/data0118/PPV_coverage/eskape_rm_plasmid_mutual_info', header = 0)
refseq_tradeoff = pd.read_csv('~/data0118/PPV_coverage/refseq_300_nrm_mutual_rm_plasmid', header = 0)
string_tradeoff = pd.read_csv('~/data0118/PPV_coverage/string_300_rm_plasmid', header = 0)
tradeoff = pd.read_csv('~/data0118/PPV_coverage/combined_rm_plasmid', header = 0)
plt.plot(tradeoff['coverage'], tradeoff['PPV'], '+', color = 'navy', label = 'PangenomeNet')
plt.plot(domain_tradeoff['coverage'], domain_tradeoff['PPV'], '+', color = 'olive', label = 'DomainNet')
plt.plot(string_tradeoff['coverage'], string_tradeoff['PPV'], '+', color = 'goldenrod', label = 'StringNet')
plt.plot(refseq_tradeoff['coverage'], refseq_tradeoff['PPV'], '+', color = 'coral', label = 'RefseqNet')
plt.plot(eskape_tradeoff['coverage'], eskape_tradeoff['PPV'], '+', color = 'crimson',  label = 'EskapeNet')

plt.legend()

plt.xlabel('coverage')
plt.ylabel('PPV')

plt.savefig('/home/hermuba/data0118/fig/PPV_coverage.pdf', format = 'pdf', dpi = 300, pad_inches = 1, bbox_inches='tight')
```
