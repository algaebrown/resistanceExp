# get sequence from those uniprot id

import xmlschema
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord
from Bio.Alphabet import IUPAC
schema = xmlschema.XMLSchema('https://www.uniprot.org/docs/uniprot.xsd')
def ID_to_seqrecord(uniprot_id, schema = schema):
    entry_dict = schema.to_dict('https://www.uniprot.org/uniprot/{}.xml'.format(uniprot_id))
    content = entry_dict['entry'][0]
    name = content['name'][0]
    seq = content['sequence']['$']
    des = content['protein']['recommendedName']['fullName']
    
    # create seq record
    
    record = SeqRecord(Seq(seq,
                       IUPAC.protein),
                   id=uniprot_id, name=name,
                   description=des)
    
    return record

all_records = []
failed = []
for i in list(all_uniprot_id):
    try:
        all_records.append(ID_to_seqrecord(i))
    except:
        failed.append(i)
