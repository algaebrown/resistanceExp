```python
import psycopg2
from Genome.context.config import config

params = config()

# test
#params['database'] = 'hermuba'


def extract_unique_qseqid(table_name):

    # connect to DB
    conn = psycopg2.connect(**params)
    cur = conn.cursor()

    # select unique qseqid
    cur.execute("""
    
    SELECT DISTINCT qseqid
    FROM {0};
    
    """
    .format(table_name))
    # close connection
    unique_qseqid = cur.fetchall()
    
    cur.close()
    conn.close()
    
    # turn list of tuple into list
    u_list = [i[0] for i in unique_qseqid]
    
    # return
    return(u_list)

```

```python
table_name = 'blastp_out_max_evalue'
refseq_qseqid = extract_unique_qseqid(table_name)
```

```python
table_name = 'eskape_blastp_out_max_evalue'
eskape_qseqid = extract_unique_qseqid(table_name)
```
