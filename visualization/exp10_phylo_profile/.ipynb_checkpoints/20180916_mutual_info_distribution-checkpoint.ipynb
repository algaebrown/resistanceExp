```python
import pandas as pd
mu = pd.read_csv("~/data0118/mutual_info/blastp_out_max_evalue_ordinary_mutual", header = 0, nrows = 1000000)
ec_mu = pd.read_csv("~/data0118/mutual_info/eskape_blastp_out_max_evalue_ordinary_mutual", header = 0, nrows = 1000000)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
fig, axes = plt.subplots(nrows =2, ncols=2, sharex = True, sharey = 'all', figsize=(10,6))
mu['mutual_info'].plot(kind = 'hist', logx = True, logy = True, ax = axes[0,0], title = 'ref_seq, mutual')
mu['nrm_mutual_info'].plot(kind = 'hist', logx = True, logy = True, ax =axes[0,1],  title = 'ref_seq, nrm' )
ec_mu['mutual_info'].plot(kind = 'hist', logx = True, logy = True,  ax =axes[1,0],title = 'entero, mutual' )
ec_mu['nrm_mutual_info'].plot(kind = 'hist', logx = True, logy = True,  ax =axes[1,1], title = 'entero, nrm')
```

```python
fig, axes = plt.subplots(nrows =1, ncols=2, sharex = True, sharey = 'all', figsize=(10,6))
mu[['mutual_info', 'nrm_mutual_info']].plot(kind = 'scatter',x = 'mutual_info', y = 'nrm_mutual_info',ax = axes[0], title = 'refseq')
ec_mu[['mutual_info', 'nrm_mutual_info']].plot(kind = 'scatter',x = 'mutual_info', y = 'nrm_mutual_info', ax = axes[1], title = 'entero')
```
