```r
library(readxl)
df <- read_excel("../../combined_aro_mic.xlsx", range = cell_cols("A:QB"))
```

```r
df[,c('cefazolin_ris')]
```

```r
start_aro = grep("3000010", colnames(df))
end_aro = grep("3004045", colnames(df))

# turn to matrix
ma <- data.matrix(df[,start_aro:end_aro])

# draw
library(gplots)
library(RColorBrewer)

hmcol = colorRampPalette(brewer.pal(9, "RdBu"))(100)
hmap <- heatmap.2(ma, labRow =df[,c('cefazolin_ris')], dendrogram = "row", Rowv = TRUE, Colv = TRUE, scale="none", trace = "none", col = hmcol, breaks = seq(-3, 3, length.out = 101))
```
