```python
# read file, save in a list

original_sdf = open('alldrug.sdf', 'r')

temp = original_sdf.read().splitlines()
original_sdf.close()


# new file
another_sdf = open('another.sdf', 'w')
```

```python
import numpy as np
random_list = np.random.normal(5,2,30000) # random numbers will be assigned to <activity>
```

```python
# change <acitivty>
i = 0
while i < len(temp):
    line = temp[i]
    print(line, file = another_sdf)
    if line == '>  <ACTIVITY>':
        #print("activity")
        i = i+1
        print(random_list[i], file = another_sdf)
    i = i+1
```
