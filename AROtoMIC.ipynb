```python
import pandas as pd
# input file "visualize ARO"
df = pd.read_excel('../visualizeARO.xlsx')
df
```

```python
mic = pd.read_excel('../PATRIC_genomes_AMR.xlsx', sheetname = 'AMR_MATRIX_MIC')
mic
```

```python
ris = pd.read_excel('../PATRIC_genomes_AMR.xlsx', sheetname = 'AMR_MATRIX_SIR')
ris
```

```python
# PATRIC_AMR genome_id datatype 
print(df['ID'].dtype)
print(ris['genome_id'].dtype)
```

```python
# Mapping to each other
combined = df.set_index('ID').join(ris.set_index('genome_id'))
```

```python
# the index of combined is already ID. ID is not a key anymore
combined.index
```

```python
combined = combined.join(mic.set_index('genome_id'), lsuffix = '_ris', rsuffix = '_mic')
combined
```

```python
combined.to_excel('../combineMICandRIStoARO.xlsx')
```

```python
# remove all-zero columns
count = combined.count(axis = 0, numeric_only = False)
remove = (count == 0)
columnName = list(remove.index)
allZero = list(remove)

remove
```

```python
print(combined.shape, "before")
for i in range(len(allZero)):
    if allZero[i] == True:
        del combined[columnName[i]]
print(combined.shape, "after")

```

```python
l = list(combined.count(axis = 0, numeric_only = False))
l.count(750)
# there are 407 genes and the remaining are mic/ris datas
```
