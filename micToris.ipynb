```python
import pandas as pd
df = pd.read_excel('../EcoliGenomes/ECgenomesWithResistanceData.xlsx')
clsi = pd.read_excel('../CLSI_m100_breakpoint.xlsx')
```

```python

for index, row in df.iterrows():
    if pd.isnull(row['SIR']) == 1:
        # check if CLSI include that criteria
        # exist organism
        # they are all E.coli and all belong to Enterobacteriae
        # exist drug
        if row['ANTIBIOTIC'] in list(clsi['drug']):
            drug = clsi['drug'].isin([row['ANTIBIOTIC']]).tolist()
            organism = clsi['family'].isin(['Enterobacteriae']).tolist()
            indexList =  [a*b for a, b in zip(drug, organism)] #.index(1)
            true_index = [i for i, v in enumerate(indexList) if v == 1] # to drbug if more than 1 index is correct
            #print(true_index)
            if len(true_index) > 1:
                Nullspecies = clsi.ix[true_index,'species'].isnull().tolist()
                if False in Nullspecies:
                    true_index.remove(true_index[Nullspecies.index(False)]) # remove those with species specification(usually samonella)
                #Nullnotspecies = clsi.ix[true_index,'not_species'].isnull().tolist()
            true_index = true_index[0]
            # apply the rules: rule index = true_index[0]; clsi.loc[true_index[0]]
            
            # if not mixture, use R-breakpoint directly
            if int(clsi.ix[true_index, 'mixture']) == 0:
                R_breakpoint = int(clsi.ix[true_index, 'R_breakpoint(>=)'])
                S_breakpoint = int(clsi.ix[true_index, 'S_breakpoint(<=)'])
            # else parse it
            else:
                R_breakpoint = int(str(clsi.ix[true_index, 'R_breakpoint(>=)']).split('/')[0])
                S_breakpoint = int(str(clsi.ix[true_index, 'S_breakpoint(<=)']).split('/')[0])
            
            # apply the rules with breakpoints   
            if bool(int(row['MIC_NUM']) >= R_breakpoint): 
                if bool(str(row['MIC_SIGN']) in [">", ">="]) or type(row['MIC_SIGN']) == float:
                    df.ix[index, 'SIR'] = 'R'
                else:
                    df.ix[index, 'SIR'] = 'cannot determine'
            elif bool(int(row['MIC_NUM']) <= S_breakpoint):
                if bool(row['MIC_SIGN'] in ["<", "<="]) or type(row['MIC_SIGN']) == float:
                        df.ix[index, 'SIR'] = 'S'
                else:
                        df.ix[index, 'SIR'] = 'cannot determine'
            else:
                df.ix[index, 'SIR'] = 'I'
          
                
        else:
            print(row['ANTIBIOTIC'], 'not exist')
df
        
```
